```python
import numpy as np
import pandas as pd
import urllib.request
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
```

```python
urllib.request.urlretrieve(url,"heart.csv")
# Retrieves Data from URL
```

```python
column_names = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang",
"oldpeak", "slope", "ca", "thal", "target"]
```

```python
df = pd.read_csv("heart.csv",names=column_names)
df
```

```python
columns_with_qm = df.columns[df.isin(["?"]).any()]
columns_with_qm
```

```python
for cols in columns_with_qm:
    mode = df[cols].mode()[0]
    df.loc[df[cols] =="?", cols] = mode    
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['ca'] = le.fit_transform(df['ca'])
```

```python
df['thal'] = le.fit_transform(df['thal'])
```

```python
df.info()
```

```python
df.describe()
```

```python
def outlier_remover(df,col):
    Q1= df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    upper_limit = Q3 + 1.5*IQR
    lower_limit = Q1 - 1.5*IQR
    df[col] = np.clip(df[col],upper_limit,lower_limit)
    return df[col]

for col in column_names:
     df[col] = outlier_remover(df,col)
```

```python
x = df.drop('target', axis=1)
y = df.target
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
sc = StandardScaler()
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

To compare Logistic Regression Model and decision tree

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
classificationReport = classification_report(y_test,y_pred)
```
