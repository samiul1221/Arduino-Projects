```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
X_data = iris.data
# To only extract the data as np array
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X_data)
```

```python
len(X)
```

```python
wcss = []
```

```python
for i in range(1,11):
    kmean = KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmean.fit(X)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss,marker='o',linestyle='--')
plt.title("Elbow Method")
plt.xlabel("No. of cluster")
plt.ylabel("WCSS")
plt.xticks(range(1,11))
plt.grid(True)
plt.show()
```

```python
kmean = KMeans(n_clusters=2)
kmean.fit(X)
```

```python
cluster_label = kmean.labels_
```

```python
cluster_label
```

```python
from sklearn.metrics import silhouette_score

ss = silhouette_score(X,cluster_label)
```

```python
ss
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=X[:,0],y=X[:,1],hue=cluster_label,palette="viridis")
plt.xlabel("Column1")
plt.ylabel("Column2")
plt.title("K-mean Clustering")
plt.legend(title="Cluster")
plt.show()

# ScatterPlot for visualizing Clustering
```
