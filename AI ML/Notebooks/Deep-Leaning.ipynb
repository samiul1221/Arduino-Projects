```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("cardekho_data.csv") # To predict Selling price
df
```

```python
df.info()
```

```python
df['Year'] = 2024 - df["Year"]
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Car_Name"] = le.fit_transform(df['Car_Name'])
df["Seller_Type"] = le.fit_transform(df["Seller_Type"])
df["Fuel_Type"] = le.fit_transform(df["Fuel_Type"])
df["Transmission"] = le.fit_transform(df["Transmission"])
df.info()
```

```python
X = df.drop("Selling_Price",axis=1)
X
```

```python
Y = df["Selling_Price"]
Y
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
# Input Layer
# Don't include activation argument if you want Linear Regression 
```

```python
model.add(Dense(64, activation='relu'))
# model.add(Dense(64, activation='relu'))
# A Hidden Layer
```

```python
model.add(Dense(1, activation='sigmoid'))
# Output Layer
```

```python
optimize_of_model  = Adam(learning_rate=0.001)
model.compile(optimizer=optimize_of_model,loss="mean_squared_error",metrics=['accuracy'])
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```
